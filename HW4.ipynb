### 1.1 Данные

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc-house-data.csv')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 4))
plt.hist(df['price'], bins = 20)
plt.title('Cтоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв.м.')
plt.ylabel('Количество');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['yr_built'], bins=20)
plt.title('Распределение года постройки')
plt.xlabel('год')
plt.ylabel('Количество');
```

## 2 задача

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(4, 4))
plt.title('Распределение домов с видом на набережную')
plt.pie(data, autopct='%1.1f%%')
plt.legend(['нет', 'да']);
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 6))
plt.title('Распределение количества этажей')
plt.pie(data, autopct='%1.1f%%', labels=data.index);
plt.legend();
```

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 6))
plt.title('Распределение состояния домов')
plt.pie(data, autopct='%1.1f%%', labels=data.index);
plt.legend();
```

##  3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Стоимость дома зависит от его жилой площади - чем она больше, тем дороже дом.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Стоимость дома зависит от его класса(качества) - чем он больше, тем дороже дом.

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5, showfliers = False);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Распределение цены по классу(качеству)');
```

Небольшую стоимость имеют дома с более низким классом, чем выше качество дома тем он дороже

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5, showfliers = False);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Распределение цены по этажности домов');
```

Меньше всего стоят дома имеющие 1 этаж. Наибольшую стоимость имеют дома с кол-вом 2,5 этажей. А в целом стоимость не сильно колеблится в зависимости от этажности.

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5, showfliers = False);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Распределение цены по виду из окон');
```

Если оценка вида больше, то стоимость дома тоже растет
